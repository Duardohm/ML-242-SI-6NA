```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
```

```python
# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]
```

```python
# Ler o arquivo CSV com as colunas especificadas
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

Tarefa 1: Análise Exploratória de Variáveis

```python
# 1.1 Contagem das diferentes classes presentes no dataset
classe_counts = vinhos['classe'].value_counts()
print(classe_counts)
```

```python
# 1.2 Estatísticas descritivas para as colunas numéricas
estatisticas = vinhos.describe()
print(estatisticas)
```

```python
# 1.3 Contagem de valores únicos para as variáveis 'classe', 'alcool', 'acido_malico', 'magnesio', 'matiz'
classe_counts = vinhos['classe'].value_counts()
alcool_describe = vinhos['alcool'].describe()
acido_malico_describe = vinhos['acido_malico'].describe()
magnesio_describe = vinhos['magnesio'].describe()
matiz_describe = vinhos['matiz'].describe()

print(classe_counts)
print(alcool_describe)
print(acido_malico_describe)
print(magnesio_describe)
print(matiz_describe)
```

# Tarefa 2: Criação de Boxplots

```python
# 2.1 Boxplot para a variável 'magnesio'
plt.figure(figsize=(8, 6))
sns.boxplot(x=vinhos['magnesio'])
plt.title('Boxplot de Magnesio')
plt.show()
```

```python
# 2.2 Boxplot para a variável 'prolina'
plt.figure(figsize=(8, 6))
sns.boxplot(x=vinhos['prolina'])
plt.title('Boxplot de Prolina')
plt.show()
```

```python
# 2.3 Boxplot para a variável 'cinzas'
plt.figure(figsize=(8, 6))
sns.boxplot(x=vinhos['cinzas'])
plt.title('Boxplot de Cinzas')
plt.show()
```

```python
# 2.4 Boxplot para a variável 'alcool'
plt.figure(figsize=(8, 6))
sns.boxplot(x=vinhos['alcool'])
plt.title('Boxplot de Álcool')
plt.show()
```

# Tarefa 3: Boxplots Segregados por Classe

```python
# 3.1 Boxplot para a variável 'magnesio' segregado por 'classe'
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot de Magnesio por Classe')
plt.suptitle('')
plt.show()
```

```python
# 3.2 Boxplot para a variável 'fenois_totais' segregado por 'classe'
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot de Fenóis Totais por Classe')
plt.suptitle('')
plt.show()
```

```python
# 3.3 Boxplot para a variável 'acido_malico' segregado por 'classe'
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot de Ácido Malico por Classe')
plt.suptitle('')
plt.show()
```

```python
# 3.4 Boxplot para a variável 'alcool' segregado por 'classe'
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot de Álcool por Classe')
plt.suptitle('')
plt.show()
```

# Tarefa 4: Criação de Histogramas

```python
# 4.1 Histograma de 'fenois_totais' com 10 bins
plt.figure(figsize=(8, 6))
plt.hist(vinhos['fenois_totais'], bins=10, edgecolor='black')
plt.title('Histograma de Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```

```python
# 4.2 Histograma de 'fenois_totais' para a classe 3, com 10 bins
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.figure(figsize=(8, 6))
plt.hist(vinhos_classe_3['fenois_totais'], bins=10, edgecolor='black')
plt.title('Histograma de Fenóis Totais - Classe 3')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```
