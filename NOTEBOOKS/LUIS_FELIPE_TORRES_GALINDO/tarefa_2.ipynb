```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# Exibindo as primeiras linhas do dataset
vinhos.head()
```

# 1. Análise Exploratória de Variáveis

```python
# Contagem das diferentes classes
classe_counts = vinhos['classe'].value_counts()
print("Contagem das classes:\n", classe_counts)

# Estatísticas descritivas das colunas solicitadas
descricao = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()
print("\nEstatísticas descritivas:\n", descricao)
```

# 2. Criação de Boxplots

```python
# Configurando o estilo do seaborn
sns.set(style="whitegrid")

# Criando boxplots
plt.figure(figsize=(12, 8))

# Boxplot para 'magnesio'
plt.subplot(2, 2, 1)
sns.boxplot(y=vinhos['magnesio'])
plt.title('Boxplot de Magnésio')
plt.ylabel('Magnésio')

# Boxplot para 'prolina'
plt.subplot(2, 2, 2)
sns.boxplot(y=vinhos['prolina'])
plt.title('Boxplot de Prolina')
plt.ylabel('Prolina')

# Boxplot para 'cinzas'
plt.subplot(2, 2, 3)
sns.boxplot(y=vinhos['cinzas'])
plt.title('Boxplot de Cinzas')
plt.ylabel('Cinzas')

# Boxplot para 'alcool'
plt.subplot(2, 2, 4)
sns.boxplot(y=vinhos['alcool'])
plt.title('Boxplot de Álcool')
plt.ylabel('Álcool')

plt.tight_layout()
plt.show()
```

# 3. Boxplots Segregados por Classe

```python
# Criando boxplots segregados por classe
plt.figure(figsize=(12, 8))

# Boxplot para 'magnesio'
plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot de Magnésio por Classe')
plt.suptitle('')
plt.ylabel('Magnésio')

# Boxplot para 'fenois_totais'
plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot de Fenóis Totais por Classe')
plt.suptitle('')
plt.ylabel('Fenóis Totais')

# Boxplot para 'acido_malico'
plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot de Ácido Málico por Classe')
plt.suptitle('')
plt.ylabel('Ácido Málico')

# Boxplot para 'alcool'
plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot de Álcool por Classe')
plt.suptitle('')
plt.ylabel('Álcool')

plt.tight_layout()
plt.show()
```

# 4. Criação de Histogramas

```python
# Histograma para 'fenois_totais'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(vinhos['fenois_totais'], bins=10, color='blue', alpha=0.7)
plt.title('Histograma de Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')

# Histograma para 'fenois_totais' filtrando a classe 3
plt.subplot(1, 2, 2)
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.hist(vinhos_classe_3['fenois_totais'], bins=10, color='green', alpha=0.7)
plt.title('Histograma de Fenóis Totais - Classe 3')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
