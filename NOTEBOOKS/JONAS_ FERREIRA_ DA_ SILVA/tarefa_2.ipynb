```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# Estatísticas Descritivas
print(vinhos['alcool'].describe())
print(vinhos['acido_malico'].describe())
print(vinhos['magnesio'].describe())
print(vinhos['matiz'].describe())

# Contagem das Classes
print(vinhos['classe'].value_counts())

# Boxplot para magnesio
sns.boxplot(y='magnesio', data=vinhos)
plt.show()

# Boxplot para prolina
sns.boxplot(y='prolina', data=vinhos)
plt.show()

# Boxplot para cinzas
sns.boxplot(y='cinzas', data=vinhos)
plt.show()

# Boxplot para alcool
sns.boxplot(y='alcool', data=vinhos)
plt.show()

# Boxplots Segregados por Classe
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Magnesio por Classe')
plt.suptitle('')  # Remove o título padrão
plt.xlabel('Classe')
plt.ylabel('Magnesio')
plt.show()

vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Fenois Totais por Classe')
plt.suptitle('')  # Remove o título padrão
plt.xlabel('Classe')
plt.ylabel('Fenois Totais')
plt.show()

vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Ácido Málico por Classe')
plt.suptitle('')  # Remove o título padrão
plt.xlabel('Classe')
plt.ylabel('Ácido Málico')
plt.show()

vinhos.boxplot(column='alcool', by='classe')
plt.title('Álcool por Classe')
plt.suptitle('')  # Remove o título padrão
plt.xlabel('Classe')
plt.ylabel('Álcool')
plt.show()

# Histograma para fenois_totais
sns.histplot(vinhos['fenois_totais'], bins=10)
plt.title('Distribuição de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()

# Histograma para fenois_totais filtrando classe 3
fenois_totais_classe_3 = vinhos[vinhos['classe'] == '3']['fenois_totais']
sns.histplot(fenois_totais_classe_3, bins=10)
plt.title('Distribuição de Fenois Totais para Classe 3')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
