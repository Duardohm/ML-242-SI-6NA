#Análise Exploratória de Dados - Vinhos (UCI)
Importando o dataset e carregando como dataframe no pandas

```python
import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

#Tarefas
#1. Análise Exploratória de Variáveis
**Realize uma análise exploratória básica das seguintes colunas, fornecendo estatísticas descritivas e insights:**

classe: Faça a contagem das diferentes classes presentes no dataset.

*   alcool
*   acido_malico
*   magnesio
*   matiz



```python
print("Visão Geral Estatística:")
classe_vision_stats = vinhos['classe'].describe()
display(classe_vision_stats)

print("\n Visão Geral:")
overview = vinhos['classe'].value_counts()
display(overview)

print("\n Colunas:")
stat = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()
display(stat)
```

#2. Criação de Boxplots
**Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:**
* magnesio
* prolina
* cinzas
* alcool

```python
import matplotlib.pyplot as plt

# Crie boxplots para as colunas especificadas
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

vinhos.boxplot(column='magnesio', ax=axes[0, 0])
axes[0,0].set_title('Magnesio')

vinhos.boxplot(column='prolina', ax=axes[0, 1])
axes[0,1].set_title('Prolina')

vinhos.boxplot(column='cinzas', ax=axes[1, 0])
axes[1,0].set_title('Cinzas')

vinhos.boxplot(column='alcool', ax=axes[1, 1])
axes[1,1].set_title('Álcool')

plt.tight_layout()
plt.show()
```

#3. Boxplots Segregados por Classe
**Crie um boxplot das seguintes colunas, segregando os dados pela variável classe:**
* magnesio
* fenois_totais
* acido_malico
* alcool

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

vinhos.boxplot(column='magnesio', by='classe', ax=axes[0, 0])
axes[0,0].set_title('Magnesio por Classe')

vinhos.boxplot(column='fenois_totais', by='classe', ax=axes[0, 1])
axes[0,1].set_title('Fenois Totais por Classe')

vinhos.boxplot(column='acido_malico', by='classe', ax=axes[1, 0])
axes[1,0].set_title('Ácido Málico por Classe')

vinhos.boxplot(column='alcool', by='classe', ax=axes[1, 1])
axes[1,1].set_title('Álcool por Classe')

plt.tight_layout()
plt.show()
```

#4. Criação de Histograma
**Crie um histograma para visualizar a distribuição dos dados na coluna:
fenois_totais, utilizando 10 bins**

```python
plt.figure(figsize=(8, 6))
plt.hist(vinhos['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.title('Histograma de Fenois Totais')
plt.show()
```

**Crie um histograma da coluna fenois_totais, filtrando apenas a classe 3, utilizando 10 bins.**

```python
classe3 = vinhos[vinhos['classe'] == '3']

plt.figure(figsize=(8, 6))
plt.hist(classe3['fenois_totais'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma - Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
