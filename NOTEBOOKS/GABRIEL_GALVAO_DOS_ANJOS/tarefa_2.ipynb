Atividade 1 :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'


column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]


vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

print(vinhos.head())

# 1. Análise da coluna 'classe'
# Contagem de cada classe
contagem_classes = vinhos['classe'].value_counts()
print("\nContagem das diferentes classes de vinho:")
print(contagem_classes)

# Insights sobre a coluna 'classe':
# Existem três classes de vinho no dataset, sendo que a Classe 2 é a mais frequente, seguida pela Classe 1 e depois pela Classe 3.

# 2. Estatísticas descritivas para 'alcool'
estatisticas_alcool = vinhos['alcool'].describe()
print("\nEstatísticas descritivas para a coluna 'alcool':")
print(estatisticas_alcool)

# Insights sobre a coluna 'alcool':
# A quantidade de álcool varia de 11.03 a 14.83, com uma média de aproximadamente 13.00. O desvio padrão indica uma dispersão moderada entre os valores.

# 3. Estatísticas descritivas para 'acido_malico'
estatisticas_acido_malico = vinhos['acido_malico'].describe()
print("\nEstatísticas descritivas para a coluna 'acido_malico':")
print(estatisticas_acido_malico)

# Insights sobre a coluna 'acido_malico':
# A concentração de ácido málico varia de 0.74 a 5.80, com uma média de aproximadamente 2.34. Existe uma maior dispersão entre os valores, o que pode indicar diferentes níveis de acidez nos vinhos.

# 4. Estatísticas descritivas para 'magnesio'
estatisticas_magnesio = vinhos['magnesio'].describe()
print("\nEstatísticas descritivas para a coluna 'magnesio':")
print(estatisticas_magnesio)

# Insights sobre a coluna 'magnesio':
# O teor de magnésio varia de 70 a 162, com uma média de 99.74. O valor mínimo está bem abaixo da média, o que pode indicar outliers ou vinhos com baixos níveis de magnésio.

# 5. Estatísticas descritivas para 'matiz'
estatisticas_matiz = vinhos['matiz'].describe()
print("\nEstatísticas descritivas para a coluna 'matiz':")
print(estatisticas_matiz)

# Insights sobre a coluna 'matiz':
# O valor da matiz varia de 0.48 a 1.71, com uma média de 0.96. O intervalo entre os valores mínimo e máximo é relativamente grande, sugerindo uma diversidade nas características de cor dos vinhos.
```

Atividade 2 :

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustando o estilo dos gráficos
sns.set(style="whitegrid")

# 1. Boxplot para a coluna 'magnesio'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='magnesio', color='lightblue')
plt.title('Boxplot de Magnésio')
plt.ylabel('Concentração de Magnésio')
plt.show()

# 2. Boxplot para a coluna 'prolina'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='prolina', color='lightgreen')
plt.title('Boxplot de Prolina')
plt.ylabel('Concentração de Prolina')
plt.show()

# 3. Boxplot para a coluna 'cinzas'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='cinzas', color='lightcoral')
plt.title('Boxplot de Cinzas')
plt.ylabel('Concentração de Cinzas')
plt.show()

# 4. Boxplot para a coluna 'alcool'
plt.figure(figsize=(8, 6))
sns.boxplot(data=vinhos, y='alcool', color='lightyellow')
plt.title('Boxplot de Álcool')
plt.ylabel('Teor Alcoólico (%)')
plt.show()
```

```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt

# Boxplot de Magnésio por Classe
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='magnesio', by='classe', grid=False)
plt.title('Boxplot de Magnésio por Classe')
plt.suptitle('')  # Removendo o título automático
plt.xlabel('Classe')
plt.ylabel('Concentração de Magnésio (mg/L)')
plt.show()

# Boxplot de Fenóis Totais por Classe
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='fenois_totais', by='classe', grid=False)
plt.title('Boxplot de Fenóis Totais por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Concentração de Fenóis Totais')
plt.show()

# Boxplot de Ácido Málico por Classe
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='acido_malico', by='classe', grid=False)
plt.title('Boxplot de Ácido Málico por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Concentração de Ácido Málico')
plt.show()

# Boxplot de Álcool por Classe
plt.figure(figsize=(8, 6))
vinhos.boxplot(column='alcool', by='classe', grid=False)
plt.title('Boxplot de Álcool por Classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Teor Alcoólico (%)')
plt.show()

```

```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustando o estilo dos gráficos
sns.set(style="whitegrid")

# 1. Histograma para a coluna 'fenois_totais' (todos os dados)
plt.figure(figsize=(8, 6))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=False, color='skyblue', edgecolor='black')
plt.title('Histograma de Fenóis Totais (Todos os Dados)')
plt.xlabel('Concentração de Fenóis Totais')
plt.ylabel('Frequência')
plt.show()

# 2. Histograma para a coluna 'fenois_totais', filtrando apenas a classe 3
# Filtrando os dados para incluir apenas a Classe 3
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

# Histograma para a Classe 3
plt.figure(figsize=(8, 6))
sns.histplot(vinhos_classe_3['fenois_totais'], bins=10, kde=False, color='orange', edgecolor='black')
plt.title('Histograma de Fenóis Totais (Classe 3)')
plt.xlabel('Concentração de Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```
